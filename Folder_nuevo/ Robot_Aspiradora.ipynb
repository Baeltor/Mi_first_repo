```
from IPython.display import display
import ipywidgets as widgets
import time

class Tablero:

  def __init__( self, tamano_celda=(50, 50) ):

    global dimension
    global posicionX
    global posicionY
    #global energia

    dimension = int( input( "digite la dimensi√≥n de la matriz " ) )
    posicionX = int( input( "digite la posici√≥n de inicio de la aspiradora en x " ) )
    posicionY = int( input( "digite la posici√≥n de inicio de la aspiradora en y " ) )
    #energia   = int( input( "Digite la energia de la aspiradora ") )

    n_celdas=( dimension, dimension )
    self.out = widgets.HTML()
    display(self.out)
    self.tamano_celda = tamano_celda
    self.n_celdas = n_celdas

  def dibujar(self, objetos):

    tablero = "<table border='1' >{}</table>"
    filas = ""

    for i in range(self.n_celdas[0]):
      s = ""
      for j in range(self.n_celdas[1]):
        contenido =""
        for o in objetos:
          if o.x == j and o.y == i:
            contenido = \
            "<div style='transform: rotate({angulo}deg);font-size:{tamano_emoticon}px;'>{emoticon}</div>".\
            format(angulo = o.angulo, tamano_emoticon = o.tamano_emoticon, emoticon = o.emoticon)
        s += "<td style='height:{alto}px;width:{ancho}px'>{contenido}</td>".\
          format(alto = self.tamano_celda[0], ancho = self.tamano_celda[1],
                contenido = contenido)
      filas += "<tr>{}</tr>".format(s)
    tablero = tablero.format(filas)
    self.out.value = tablero
```

```
class Agente:

  def __init__(self, x=any, y=any, angulo=0, emoticon="ü§ñ", tamano_emoticon=30, energia=any, aspiro=0, recorrio=0):

    self.energia         = energia
    self.aspiro          = aspiro
    self.recorrio        = 0
    self.x               = x
    self.y               = y
    self.angulo          = angulo
    self.emoticon        = emoticon
    self.tamano_emoticon = tamano_emoticon
    self.direccionY      = "abajo"
    self.direccionX      = "arriba"

  def recorrer(self, objetos):

    if self.energia > 0:

      for i in range(dimension):

        if self.direccionY == "abajo":

          if self.y == ( dimension - 1 ) and self.x == ( dimension - 1 ):

            self.x = 0
            self.y = 0
            self.recorrio += 1
            self.energia -= 1
            time.sleep(1)
            break

          if self.y == 0:

            self.y += 1
            self.recorrio += 1
            self.energia -= 1
            time.sleep(1)
            break

          if self.y > 0 and self.y < ( dimension - 1 ):

            self.y += 1
            self.recorrio += 1
            self.energia -= 1
            time.sleep(1)
            break

          if self.y == ( dimension - 1 ):

            self.x += 1
            self.recorrio += 1
            self.energia -= 1
            self.direccionY = "arriba"
            time.sleep(1)
            break


        if self.direccionY == "arriba":

          if self.x == ( dimension - 1 ) and self.y == 0:

            self.x = 0
            self.y = 0
            self.recorrio += 1
            self.energia -= 1
            self.direccionY = "abajo"
            time.sleep(1)
            break

          if self.y == 0:

            self.direccionY = "abajo"
            self.x += 1
            self.recorrio += 1
            self.energia -= 1
            time.sleep(1)
            break

          if self.y < ( dimension - 1 ) and self.y > 0:

            self.y -= 1
            self.recorrio += 1
            self.energia -= 1
            time.sleep(1)
            break

          if self.y == ( dimension - 1 ):

            self.y -= 1
            self.recorrio += 1
            self.energia -= 1
            time.sleep(1)
            break

      for k, objeto in enumerate( objetos ):

        if self.y == objeto.y and self.x == objeto.x and self.energia > 0:

          self.energia  -= 1
          self.aspiro   += 1
          self.recorrio += 1
          del objetos[k]
          break

  def aspirar(self, objetos):

    countObjetos = ( len( objetos ) - 1 )

    if self.energia > 0:

      for k, objeto in enumerate( objetos ):

        if k < (countObjetos - 1):

          if self.direccionY == "abajo":

            estadoAbajo = True

            for objetoTemp in objetos:

              if ( self.y + 1 ) == objetoTemp.y and self.x == objetoTemp.x:

                estadoAbajo = False

            if estadoAbajo:

              if self.y == objeto.y and ( self.x + 1 == objeto.x ) and self.energia > 0:

                if( self.x <= dimension ):

                  self.x += 1
                  self.energia  -= 1
                  self.aspiro   += 1
                  self.recorrio += 1
                  del objetos[k]
                  time.sleep(1)
                  break

              if self.y == objeto.y and ( self.x - 1 == objeto.x ) and self.energia > 0:

                if( self.x <= dimension ):

                  self.x -= 1
                  self.energia  -= 1
                  self.aspiro   += 1
                  self.recorrio += 1
                  del objetos[k]
                  time.sleep(1)
                  break

          if self.direccionY == "arriba":

            estadoArriba = True

            for objetoTemp2 in objetos:

              if ( self.y - 1 ) == objetoTemp2.y and self.x == objetoTemp2.x:

                estadoArriba = False

            if estadoArriba:

              if self.y == objeto.y and ( self.x + 1 == objeto.x ) and self.energia > 0:

                if( self.x <= dimension ):

                  self.x += 1
                  self.energia  -= 1
                  self.aspiro   += 1
                  self.recorrio += 1
                  del objetos[k]
                  time.sleep(1)
                  break

              if self.y == objeto.y and ( self.x - 1 == objeto.x ) and self.energia > 0:

                if( self.x <= dimension ):

                  self.x -= 1
                  self.energia  -= 1
                  self.aspiro   += 1
                  self.recorrio += 1
                  del objetos[k]
                  time.sleep(1)
                  break
```
